# Programs using matplotlib /  seaborn for data visualisation
a)Write a program to draw univariate visualization plots(line plot,histogram,boxplot,barchart,piechart) with matplotlib for iris dataset

b)Write a program to draw multivariate visualization plots(scatter plots, scatter multiple,scatter matrix,bubble plot) with matplotlib for iris dataset

c)Write a program to draw univariate and multivariate visualization plots with seaborn(line plot, pairplot,jointplot,heatmap) for iris dataset


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
df=pd.read_csv("iris.csv")
df=pd.DataFrame(df)
x=df["sepal_length"]
plt.hist(x,bins=20,color="green")
plt.title("Sepal Length")
plt.xlabel("Sepal_Length")
plt.ylabel("Count")

df.plot(kind="hist")

#HISTOGRAM AND LINE




```

```python
df.plot(kind="line")

df.plot(kind="line",subplots=True)
```

```python
df.boxplot() #BOX chart
```

```python
df["sepal_length"].plot(kind = 'box')
#Box chart

```

```python
df["sepal_length"].plot.bar() #BAR PLOT
```

```python
plt.pie(df['species'].value_counts())
```

```python
plt.pie(df['species'].value_counts(),labels=['setosa','versicolor','virginica'],shadow=True) #PIE CHART
```

# QB

```python
df.plot(kind="scatter",x="sepal_length",y="petal_width") #SCATTER PLOT
```

```python
ax1 = df.plot(kind='scatter', x='petal_length', y = 'petal_width', label='petal_length Vs petal_width', color='r')    
ax2 = df.plot(kind='scatter', x='petal_length', y = 'sepal_length', label='petal_length Vs sepal_length',color='g',ax=ax1)    
ax3 = df.plot(kind='scatter',  x='petal_length', y = 'sepal_width',label='petal_length Vs sepal_width', color='b',ax=ax2)
print(ax1 == ax2 == ax3) 
#MULTIPLE SCATTER PLTS
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df,alpha=0.8,diagonal='kde')
#SCATTER MATRIX
```

```python
plt.scatter('petal_length', 'petal_width',  s='sepal_length',  alpha=0.5,  data=df) #BUBBLE PLOT
```

# Q3

```python
sns.lineplot(data=df) #LINE PLOT
```

```python
sns.pairplot(data=df,kind="scatter") #PAIR PLOT
```

```python
sns.jointplot(x="sepal_length",y="sepal_width",data=df,height=8) #JOINT PLOT
```

```python
sns.boxplot(data=df) #BOXPLOT
```

```python
sns.heatmap(df.corr(),annot=True) #HEAT MAP
```
